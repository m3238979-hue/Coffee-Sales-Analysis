```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"C:\Users\hp\Downloads\Coffee sales in a vending machine.csv")
```

```python
df.head()  #Quick view of data
```

```python
df.info() # Check data types and missing values
```

```python
df["coffee_name"].value_counts()
```

```python
# Total revenue for each type of coffee
df.groupby("coffee_name")["money"].sum().sort_values(ascending = False)
```

```python
# Payment methods distribution account
df["cash_type"].value_counts()
```

```python
# Best selling product
sns.countplot(data=df,x="coffee_name")
plt.xticks(rotation=45)
plt.title("best-selling product")
plt.show()
```

```python
# Revenue by coffee type and payment method
sns.barplot(data=df,x="coffee_name",y="money",hue="cash_type")
plt.xticks(rotation=45)
plt.show()
```

```python
# Money Distribution by paymemt type
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x="cash_type",y="money")
plt.title("Money Distribution by payment type")
plt.show()
```

```python
# Most frequent customers
top_customers = df["card"].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_customers.index, y=top_customers.values, palette="rocket",)
plt.xticks(rotation=90)
plt.title("Most frequent customers")
plt.show()
```
